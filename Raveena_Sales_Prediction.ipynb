```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Simple dataset
data = {
    'advertising_budget': [1000, 2000, 3000, 4000, 5000],
    'sales': [15000, 25000, 35000, 45000, 55000]
}

df = pd.DataFrame(data)

# Features & target
X = df[['advertising_budget']]
y = df['sales']

# Model
model = LinearRegression()
model.fit(X, y)

# Predict sales for new budget
prediction = model.predict([[6000]])

print("Predicted Sales for 6000 budget:", prediction)
```
